# K-means

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset, membership, centers = datasets.make_blobs(n_samples=500, n_features=2,
                                                   centers=3, return_centers=True,
                                                   cluster_std=1.0)
```

```python
plt.scatter(dataset[(membership==0),0], dataset[(membership==0),1], color='r')
plt.scatter(dataset[(membership==1),0], dataset[(membership==1),1], color='g')
plt.scatter(dataset[(membership==2),0], dataset[(membership==2),1], color='b')
plt.show()
```

```python
k=3
```

```python
k0 = np.random.uniform(low=-10.0, high=10.0, size=2)
k1 = np.random.uniform(low=-10.0, high=10.0, size=2)
```

```python
n_data = dataset.shape[0]
```

```python
our_membership = np.zeros([n_data], dtype=int)
```

```python
movement_0 = np.inf
movement_1 = np.inf
```

```python
while (movement_0>0.0001 and movement_1>0.0001):
    
    print('k0={} k1={}'.format(k0, k1))
    
    for ii in range(n_data):
        coors = dataset[ii,:]
        dist0 = np.sqrt( (coors[0]-k0[0])**2 +  (coors[1]-k0[1])**2 )
        dist1 = np.sqrt( (coors[0]-k1[0])**2 +  (coors[1]-k1[1])**2 )
        if dist0 < dist1:
            our_membership[ii]=0
        else:
            our_membership[ii]=1
        
    mask = (our_membership == 0)
    new_k0 = [np.mean(dataset[mask, 0]), np.mean(dataset[mask, 1])]
    
    mask = (our_membership == 1)
    new_k1 = [np.mean(dataset[mask, 0]), np.mean(dataset[mask, 1])]
    
    movement_0 = np.sqrt( (new_k0[0]-k0[0])**2 +  (new_k0[1]-k0[1])**2 )
    movement_1 = np.sqrt( (new_k1[0]-k1[0])**2 +  (new_k1[1]-k1[1])**2 )
    
    print('new_k0={} new_k1={}'.format(new_k0, new_k1))
    print(movement_0, movement_1)
    print(' ')
    
    k0 = new_k0
    k1 = new_k1
```

```python
plt.scatter(dataset[(our_membership==0),0], dataset[(our_membership==0),1], color='r')
plt.scatter(dataset[(our_membership==1),0], dataset[(our_membership==1),1], color='g')
plt.show()
```
